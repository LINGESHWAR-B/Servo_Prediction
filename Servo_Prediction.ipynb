**What is Servo:**
A servo motor is defined as an electric motor that provides precise control of angular or linear position, speed, and torque using a feedback loop system.
Servo motors are crucial in high-precision fields such as robotics, CNC machinery, and automated manufacturing for their ability to handle complex motions and tasks.

**Get Categories and Counts of Categorical Variables**

```python
df[['Motor']].value_counts()
```

```python
df[['Screw']].value_counts()
```

**Get Encoding of Categorical Features**

```python
df.replace({'Motor':{'A':0, 'B':1, 'C':2, 'D':3, 'E':4}}, inplace = True)
```

```python
df.replace({'Screw':{'A':0, 'B':1, 'C':2, 'D':3, 'E':4}}, inplace = True)
```

**Define y (dependent or label or target variable) and X (independent or features or attribute Variable)**

```python
y = df['Class']
```

```python
y.shape
```

```python
y
```

```python
X = df[['Motor', 'Screw', 'Pgain', 'Vgain']]
```

```python
X = df.drop('Class', axis = 1)
```

```python
X
```

**Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Get Model Train**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

**Get Model Prediction**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

**Get Model Evaluation**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

**Get Visualization of Actual Vs Predicted Results**

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
